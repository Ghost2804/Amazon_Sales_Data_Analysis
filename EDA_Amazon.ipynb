```python
def analysis(df):

    df.plot(kind='hist', bins=20)
    plt.title('Histogram')
    plt.show()
    
    df.plot(kind='kde')
    plt.title('KDE Plot')
    plt.show()
   
    df.plot(kind='box')
    plt.title('Box Plot')
    plt.show()
    
    print(df.describe())

```

```python
analysis(dt.Unit_Price)
```

```python
def char_analysis(dt):
    dt.value_counts().nlargest(20).plot(kind = 'pie', autopct = '%0.1f%%')
    plt.title('Piechart')
    plt.show()

    dt.value_counts().nlargest(20).plot(kind= 'bar')
    plt.title("bar")
    plt.show()
```

```python
char_analysis(dt.Region)
```

```python
char_analysis(dt.Country)
```

```python
char_analysis(dt.Item_Type)
```

```python
char_analysis(dt.Order_Priority)
```

```python
char_analysis(dt.Sales_Channel)
```

```python
dt['Year_Month'] = dt.Order_Date.dt.to_period('M')
```

```python

time_series_data = dt.groupby('Year_Month').agg({'Total_Revenue':'sum', 'Total_Profit': 'sum'}).reset_index()
```

```python
time_series_data.head(5)
```

```python
plt.figure(figsize=(25,10))
plt.plot(time_series_data['Year_Month'].head(27).astype(str), time_series_data['Total_Revenue'].head(27), label = 'Total Revenue')
plt.plot(time_series_data['Year_Month'].head(27).astype(str), time_series_data['Total_Profit'].head(27), label = 'Total_Profit')
plt.xlabel('Year-Month')
plt.ylabel('Amount')
plt.title('Sales and Profit Over Time')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(25,10))
plt.plot(time_series_data['Year_Month'].tail(27).astype(str), time_series_data['Total_Revenue'].tail(27), label = 'Total Revenue')
plt.plot(time_series_data['Year_Month'].tail(27).astype(str), time_series_data['Total_Profit'].tail(27), label = 'Total_Profit')
plt.xlabel('Year-Month')
plt.ylabel('Amount')
plt.title('Sales and Profit Over Time')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(25,10))
plt.plot(time_series_data['Year_Month'].astype(str), time_series_data['Total_Revenue'], label = 'Total Revenue')
plt.plot(time_series_data['Year_Month'].astype(str), time_series_data['Total_Profit'], label = 'Total_Profit')
plt.xlabel('Year-Month')
plt.ylabel('Amount')
plt.title('Sales and Profit Over Time')
plt.legend()
plt.show()
```

```python


dt['Shipping_Time'] = (dt['Ship_Date'] - dt['Order_Date']).dt.days


plt.figure(figsize=(10, 6))
sns.histplot(dt['Shipping_Time'], bins=20, kde=True, color='purple')
plt.title('Shipping Time Distribution')
plt.xlabel('Days')
plt.ylabel('Frequency')
plt.show()


plt.figure(figsize=(10, 7))
sns.boxplot(x='Region', y='Shipping_Time', data=dt)
plt.title('Shipping Time by Region')
plt.xlabel('Region')
plt.ylabel('Shipping Time (Days)')
plt.xticks(rotation=45)
plt.show()
```
